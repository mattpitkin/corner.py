```python
import corner
import numpy as np
import matplotlib.pyplot as pl
```

```python
x = np.random.randn(50000, 2)
```

```python
fig = corner.corner(x, quantiles=(0.16, 0.84), levels=(1-np.exp(-0.5),))
fig.suptitle("correct `sigma' levels", fontsize=22);
```

```python
fig = corner.corner(x, quantiles=(0.16, 0.84), levels=(0.68,))
fig.suptitle("incorrect `sigma' levels", fontsize=22);
```
